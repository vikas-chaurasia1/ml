```python
import cv2
cap=cv2.VideoCapture(0)      #here 0 by default web cam id
while True:
    ret,frame=cap.read()
    gray_frame=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    if ret==False:
        continue
    cv2.imshow("hi chaurasia",frame)
    cv2.imshow("hey gresih",gray_frame)
    #wait user input  q for quit
    quit=cv2.waitKey(1) & 0xFF
    if quit==ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```

```python
#video sreaming through  harcascade classfier
import cv2
cap=cv2.VideoCapture(0)
while True:
    ret,frame=cap.read()
    harcascade_object=cv2.CascadeClassifier('haarCascade.xml')
    if ret==False:
        continue
    faces=harcascade_object.detectMultiScale(frame,1.3,5)
    for (x,y,w,h) in faces:
        cv2.rectangle(frame,(x,y),(x+w,y+h),(255,0,0),2)
    cv2.imshow('selfie',frame)
    quit=cv2.waitKey(1) & 0xFF
    if quit==ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
