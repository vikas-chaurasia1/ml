```python
from sklearn.linear_model import LinearRegression
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data=pd.read_csv('grad.csv')
print(data)
print(data.shape)
data=data.values
x=data[:,1:8]
y=data[:,8:]
print(x)
print(x.shape)
print(y[1:5,:])
print(y.shape)
model=LinearRegression()
model.fit(x,y)
theta=model.coef_
cept=model.intercept_
print(theta)
print(cept)
print("hoga")
print(theta.shape)
print(cept.shape)
theta=np.array(theta)
X=np.array([324,107,4,4.5,4.5,8.87,1])
result=np.dot(theta,X)+np.array(cept)
print(theta)
print(result)
```
