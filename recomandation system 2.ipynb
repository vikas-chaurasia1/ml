```python
#col_names=["userid","itemid","rating","time"]
ratings=pd.read_csv("ratings.csv")
ratings.head(n=7)
```

```python
movies_title=pd.read_csv("movies.csv",encoding="Latin1")
movies_title.head(n=5)
```

```python
movies_data=pd.merge(ratings,movies_title,on='movieId')
movies_data.head(n=6)
```

```python
final_movies_data=movies_data.pivot_table(index='userId',columns='title',values='rating')
final_movies_data.head(n=10)
```

```python
test_movie_id=final_movies_data['101 Dalmatians (1996)']
print(test_movie_id)
similar_movies=final_movies_data.corrwith(test_movie_id)
recomd=pd.DataFrame(similar_movies,columns=['pearson'])
recomd.dropna(inplace=True)
recomd=recomd.sort_values('pearson',ascending=False)
movies_data1=pd.merge(recomd,movies_title,on='title')
movies_data1.head(n=10)
```
